```python
def get_id(Player):
#    if Player in NonUniqueNames['NAME'].values:
#        return -1
    IDList = playerIDs.index[playerIDs['NAME'] == Player].to_list()
    if (len(IDList) == 0):
        return -2
    if (len(IDList) == 1):
        return playerIDs['PLAYER_ID'][IDList[0]]
    if (len(IDList) > 1):
        return -1
```

```python
player_id_name = pd.read_csv("player_id_name.csv")
season_counting_stats = pd.read_csv("../season_counting_stats.csv")
season_counting_stats['TEAM_ABBREVIATION'] = season_counting_stats['TEAM_ABBREVIATION'].replace({'SAN': 'SAS', 'PHX': 'PHO'})
def disambiguate(row):
    Player = row['player_name']
    IDList = player_id_name.index[player_id_name['NAME'] == Player].to_list()
    if len(IDList) == 1:
        return player_id_name['PLAYER_ID'][IDList[0]]
    SeasonStart = row['season_start']
    Team = row['team_abv']
    #print(Player, " ", SeasonStart, " ", Team)
    IDList = season_counting_stats.index[(season_counting_stats['NAME'] == Player) & (season_counting_stats['SEASON_START'] == SeasonStart)].to_list()
    #print("With just name and year: ", Player, " ", SeasonStart, " ", IDList)
    if len(IDList) == 1:
        return season_counting_stats['PLAYER_ID'][IDList[0]]
    if len(IDList) > 1:
        IDList = season_counting_stats.index[(season_counting_stats['NAME'] == Player) & (season_counting_stats['SEASON_START'] == SeasonStart) & (season_counting_stats['TEAM_ABBREVIATION'] == Team)].to_list()
        #print("With name, year, and season", Player, " ", SeasonStart, " ", IDList)
        if len(IDList) == 1:
            return season_counting_stats['PLAYER_ID'][IDList[0]]
        if len(IDList) > 1:
            return -1
        if len(IDList) == 0:
            return -3
    if len(IDList) == 0:
        return -3
```

```python
unmatched_waivers['last_name'] = unmatched_waivers['player_name'].apply(lambda x: x.split()[-1])
player_id_name['last_name'] = player_id_name['NAME'].apply(lambda x: x.split()[-1])
```

```python
potential_matches = unmatched_waivers.merge(player_id_name, on='last_name', how='left')
```

```python
potential_matches
```

```python
single_match = potential_matches.groupby('last_name').filter(lambda x: (x['player_name'].nunique() == 1) and (x['NAME'].nunique()==1))
```

```python
unique_matches = single_match.drop_duplicates(subset='player_name', keep='first')
```

```python
waived_data.loc[waived_data['player_id'] == -3, 'player_id'] = waived_data['player_name'].map(unique_matches.set_index('player_name')['PLAYER_ID'])
```
